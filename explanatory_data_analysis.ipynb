# Gather

## Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import geopandas as gpd
import json
import os
from shapely.geometry import Point, Polygon
from geopandas.tools import sjoin
```

## Data

```python
original_data = pd.read_csv('green_tripdata_2016-02.csv')
clean_data = pd.read_csv('green_taxi_clean.csv')
df = clean_data.copy()
```

# Data Preparation

- convert time columns from object to datetime type.

```python
df['pickup_time'] = pd.to_datetime(df['pickup_time'])
df['dropoff_time'] = pd.to_datetime(df['dropoff_time'])
```

- remove pickup trips out of NYC by removing the coordinates lie out of NYC map.

```python
latitude_max = 40.93
latitude_min = 40.5
longitude_min = -74.3
longitude_max = -73.68

indices_max_longitude = df[df['pickup_longitude'] > longitude_max].index.tolist()
indices_min_longitude = df[df['pickup_longitude'] < longitude_min].index.tolist()
indices_max_latitude = df[df['pickup_latitude'] > latitude_max].index.tolist()
indices_min_latitude = df[df['pickup_latitude'] < latitude_min].index.tolist()

indices_to_drop = list(set(indices_max_latitude + indices_max_longitude + indices_min_latitude + indices_min_longitude))
df.drop(index=indices_to_drop, inplace=True)
```

## How is the trip distance related to the time of the day?

```python
mean_distance_hour_weekday = df[df['weekend']==0].groupby('hour').trip_distance.mean()
median_distance_hour_weekday = df[df['weekend']==0].groupby('hour').trip_distance.median()
mean_distance_hour_weekend = df[df['weekend']==1].groupby('hour').trip_distance.mean()
median_distance_hour_weekend = df[df['weekend']==1].groupby('hour').trip_distance.median()

plt.figure(figsize=(10, 6))
plt.plot(mean_distance_hour_weekday.index, mean_distance_hour_weekday.values, label='Weekday Mean distanceance', c='#377eb8')
plt.plot(median_distance_hour_weekday.index, median_distance_hour_weekday.values, label='Weekday Median distanceance', c='#ff7f00')
plt.plot(mean_distance_hour_weekend.index, mean_distance_hour_weekend.values, label='Weekend Mean distanceance', c='#4daf4a')
plt.plot(median_distance_hour_weekend.index, median_distance_hour_weekend.values, label='Weekend Median distanceance', c='#f781bf')
plt.ylim(1.5, 5)
plt.title('distance VS Hour', size=20)
plt.xlabel('Hour', size=15)
plt.ylabel('Trip Distance (mile)', size=15)
plt.legend()
plt.show()
```

The x-axis of this graph shows the 24 hours of the day while the distance of trips are on the y-axis. It is seen clearly that distances rose sharply at around 4:00 a.m. in the morning, then it reached the peak at around 5:00, and dropped again sharply till it reached 8:00 a.m.

The weekends graphs are shifted to the right of the weekdays.

## What are the most popular pickup locations on weekend vs weekdays?

```python
crs = {'init': 'epsg:4326'}
geometry = [Point(xy) for xy in zip(df['pickup_longitude'], df['pickup_latitude'])]

geo_df = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
geo_df.head()
```

```python
if not os.path.exists('nyc_neighbourhoods.json'):
    url = 'https://data-beta-nyc-files.s3.amazonaws.com/resources/35dd04fb-81b3-479b-a074-a27a37888ce7/d085e2f8d0b54d4590b1e7d1f35594c1pediacitiesnycneighborhoods.geojson?Signature=dlmy%2FmXtBTTjAPv9%2BVlC%2Bb%2FifrI%3D&Expires=1611459876&AWSAccessKeyId=AKIAWM5UKMRH2KITC3QA'
    neighbourhoods_info = requests.get(url).json()

    with open('nyc_neighbourhoods.json', 'w') as fp:
        json.dump(neighbourhoods_info, fp)
    
neighbourhood_list = []
with open('nyc_neighbourhoods.json') as f:
    neighbourhoods_info = json.load(f)
number_neighbourhoods = len(neighbourhoods_info.get('features'))
for i in range(number_neighbourhoods):
    neighbourhood = neighbourhoods_info.get('features')[i].get('properties').get('neighborhood')
    polygon = neighbourhoods_info.get('features')[i].get('geometry').get('coordinates')[0]
    neighbourhood_list.append({'neighbourhood':neighbourhood,
                               'polygon':polygon})

neighbourhood_df = pd.DataFrame(neighbourhood_list, columns=['neighbourhood', 'polygon'])
neighbourhood_df.polygon = neighbourhood_df.polygon.apply(lambda x: Polygon(x))
neighbourhood_df.rename(columns={'polygon':'geometry'})
neighbourhood_geo_df = gpd.GeoDataFrame(neighbourhood_df)
neighbourhood_geo_df.set_geometry('polygon', inplace=True)
```

```python
pickups_neighbourhoods = sjoin(geo_df, neighbourhood_geo_df, how='left')
pickups_neighbourhoods.dropna(inplace=True)
most_popular_location_weekend = pickups_neighbourhoods[pickups_neighbourhoods['weekend'] == 1]\
                                                .neighbourhood.value_counts().head(10)
most_popular_location_weekday = pickups_neighbourhoods[pickups_neighbourhoods['weekend'] == 0]\
                                                .neighbourhood.value_counts().head(10)
```

```python
weekend_location_df = neighbourhood_geo_df[neighbourhood_geo_df.neighbourhood.isin(most_popular_location_weekend.index)]
weekday_location_df = neighbourhood_geo_df[neighbourhood_geo_df.neighbourhood.isin(most_popular_location_weekday.index)]

fig, ax = plt.subplots(figsize=(8, 8))
neighbourhood_geo_df.plot(ax=ax, alpha=1, color='white', edgecolor='black', label='NYC')
weekend_location_df.plot(ax=ax, color='b', alpha=0.5, label='Weekend')
weekday_location_df.plot(ax=ax, color='r', alpha=0.5, label='Weekday')
plt.title('Pickup location_s', size=20)
plt.xlabel('Longitude', size=15)
plt.ylabel('Latitude', size=15)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
neighbourhood_geo_df.plot(ax=ax, alpha=1, color='white', edgecolor='black')
geo_df.plot(ax=ax, markersize=1, alpha=0.05, edgecolors='none')
plt.title('Pickup Locations Distribution', size=20)
plt.xlabel('Longitude', size=15)
plt.ylabel('Latitude', size=15)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.barh(most_popular_location_weekend.index, most_popular_location_weekend.values)
plt.gca().invert_yaxis()
plt.title('Most Popular Pickup Locations on Weekends', size=15)
plt.xlabel('Number of Pickups', size=12)
plt.ylabel('Neighbourhoods', size=12)
plt.show()
```
